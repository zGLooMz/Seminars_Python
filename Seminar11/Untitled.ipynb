```python
from sympy import *
```

1. Определить корни

```python
x = Symbol('x')
func=-5*x**2+10*x-150
y=solve(func)
x1=(y[0])
x2=(y[1])
print(x1.evalf(),x2.evalf())
```

2. Найти интервалы, на которых функция возрастает

```python
fd=diff(func)
print(solve(0<fd))
```

3. Найти интервалы, на которых функция убывает

```python
print(solve(fd<0))
```

4. Построить график

```python
import matplotlib.pyplot as plt

plot(func, fd)
```

5. Вычислить вершину

```python
corni=solve(fd)
top=corni[0]
x=top
y=-5*x**2+10*x-150
print(top,y)
```
